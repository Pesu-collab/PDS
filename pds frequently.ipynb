```python
# =========================================
# 🔥 ALL-IN-ONE PYTHON PRACTICE CODE
# =========================================

import math

# =========================================
# 1. Fibonacci Series
# =========================================
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    
    series = [0, 1]
    for i in range(2, n):
        series.append(series[-1] + series[-2])
    return series


# =========================================
# 2. Prime Number Checker
# =========================================
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True


# =========================================
# 3. Palindrome Checker
# =========================================
def is_palindrome(text):
    text = str(text)
    return text == text[::-1]


# =========================================
# 4. Word Count
# =========================================
def word_count(text):
    return len(text.split())


# =========================================
# 5. Total Value
# =========================================
def total_value():
    numbers = [10, 20, 30, 40]
    return sum(numbers)


# =========================================
# 6. Happy Number
# =========================================
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(d)**2 for d in str(n))
    return n == 1


# =========================================
# 🚀 RUN EVERYTHING (NO _name_ BLOCK)
# =========================================

n = 10

print("Fibonacci Series:", fibonacci(n))

num = 7
print("Is Prime:", num, "->", is_prime(num))

text = "madam"
print("Is Palindrome:", text, "->", is_palindrome(text))

sentence = "hello hello world"
print("Word Count:", word_count(sentence))

print("Total Value:", total_value())

happy_num = 19
```
